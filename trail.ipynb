```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
# to count hashable item from collection 
from collections import Counter
#Exercise 0
# Fetch data set from API
iris_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

# We need colomn names to make dataset easier to index (For example: SLice, dice and move through)
col_name = ["Sepal_length", "Sepal_Width", "Pedal_Length", "Pedal_Width", "Class"] 

#Read the iris dataset into dataframe 
iris = pd.read_csv(iris_url, names=col_name)
#Print fist five values
iris.head()
```

```python
#Print last five data points
iris.tail()
```

```python
#print the data types of the data
iris.dtypes
```

```python
# Excercise 1:
# 1.Find all the iris present in dataset 
print(Counter(iris["Class"]))
# 2.Replace each iris class with a numerical value
iris.replace({'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3})
```
